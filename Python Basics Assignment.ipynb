2. What is an interpreter in Python?
    - A python interpreter is a program that directly executes python code. When you write python code interpreter compiles it into bytecode. This bytecode is executed by Python Virtual Machine(PVM).  
    An interpreter translates and executes code line by line, during runtime. Its speed is slower because translation happens during execution. It stops at the first error and shows it immediately.

3. What are pre-defined keywords in Python?
    - Pre-defined keywords in python are reserved words which have specific meaning and specific purpose. These keywords are fundamental to defining logic in python program. These keywords can not be used as variable names.   
    Examples : print, type, class, def, if, elif, etc

4. Can keywords be used as variable names?
    - No, keywords can not be used as variable names because they have specific, predefined meaning. And if we assing them as a variable name it will give syntax error, because their special meaning would conflict with its use as variable name.

5. What is mutability in Python?
    - In python, mutability means that object's state or value can be changed after they are created. These objects are called mutable objects. Mutability allows operations like adding new elements, remove elements or modify existing elements.  
    Examples: list, set, dict

6. Why are lists mutable, but tuples are immutable?
    - Lists are mutable in python because they are built in such way where elements need to be added, removed, or modified after creation. Mutability allows to add, remove or modify elements in list, which makes lists more flexible.

    - Tuples are immutable in python because their fixed nature provides guarantee that the data will remain unchanged. Once created the data can't be modify. This quality of tuples make them safe for data which we do not want to be changed.

7. What is the difference between “==” and “is” operators in Python?
    - '==' is a Comparison Operator, that compares the value of two objects. It checks whether the data represented by objects is same or not. If the values are same it returns True, if not then False.

    - Whereas 'is' is a Identity Operator, that compares the identity of two objects. It checks whether the two objects/variables refer to the same location in memory or not. If they refer to the same location then it returns True, If not then False.

8. What are logical operators in Python?
    - Logical operators in python are used to combine conditional statements and perform logical operations. Logical operators returns True or False based on conditions.  
    There are three types of logical operators in python-  
    1. and operator  
        - returns True if both operands are true.
        - returns False if either or both operands are false.
    2. or operator  
        - returns True if at least one operand is true.
        - returns False if both operands are false.
    3. not operator  
        - this is a unary operator, it operates on a single operand.
        - it reverse the logical state of its operand. If the operand is True, not operator returns False, and if the operand is False, it returns True.

9. What is type casting in Python?
    - Type casting in python, is a process of converting a variable or value from one data type to another. Type casting is also known as type conversion. This is often very necessary when performing operations that requires specific data types.  
    There are two types of type casting in python -
    1. Implicit type casting
    2. Explicit type casting

10. What is the difference between implicit and explicit type casting?
    -
    - **Implicit type casting :** This occurs automatically by the Python interpreter. Python automatically converts one data type to anothor to ensure the operation can proceed.

    - **Explicit type casting :** When the programmer converts a data type to another, its called Explicit type coversion/casting. Python provides built-in functions for Explicit type casting such as: int(), float(), str(), list(), ect.

11. What is the purpose of conditional statements in Python?
    - Conditional statements control the flow of a program based on specific conditions. That enables program to make dicisions and execute different blocks of code depending on whether a given condition is True or False.  
  - Conditional statements are -   
    if statements  
    if-elif statements  
    if-elif-else statements  
    nested-if-else statements

12. How does the elif statement work?
    - The elif statement is a conditional control flow statement in python. It allows multiple conditions in a sequential manner and provides an alternative path when if condition is not met.   
 - How it works -
    - The program first check the if statement.  
    If the if condition is True, the code block within the if statement is executed.   
    - If the if statement is False, the program moves to the first elif statement.
    If the elif condition is True, the code block within the elif statement is executed.
    - If the elif statement is False, then the program moves to the next elif condition(if there is any).
    - If the above all conditions are False, the code Within the else condition will be executed.

13. What is the difference between for and while loops?
    - **for loop**   
    for loop is used when you know the number of iterations in advance. for loop is generally used for iterating over sequences like list, string or range. for loop have a known count of iteration.

    - **while loop**  
    while loop is used when the number of iterations is unknown and depends on a condition. while loop repeats the block of code as long as the given condition is true. in while loop the condition is checked before each iteration, if True, the code block executes and the process repeats untill the condition become False.

14. Describe a scenario where a while loop is more suitable than a for loop?
    - while loop is more suitable than a for loop in scenarios where the number of iteration is not known and the loop needs to continue until a specific condition is met.  
  - **Scenario -**  
     Consider a program that requires the user to enter a valid email address. The program needs to repeatedly prompt the user for input until a correctly formatted email address is provided. The number of attempts for user is unknown.  
     - In this scenario the condition is to provide a correct email address. So untill the user provide a correct email address, the program will keep asking for user input. So while loop is ideal for this situation as we have a condition but we don't know how many attampts the user have.

```python
# Que.6 Write a function to demonstrate the behavior of mutable and immutable arguments.

def modify_arguments(mutable_list, immutable_int):
  print("Inside function (before modification):")
  print("Mutable list:", mutable_list)
  print("Immutable int:", immutable_int)

  mutable_list.append("new_item")  # Modifying the list
  immutable_int += 1              # Attempting to modify the integer

  print("Inside function (after modification):")
  print("Mutable list:", mutable_list)
  print("Immutable int:", immutable_int)

# Demonstrate with a list (mutable) and an integer (immutable)
my_list = [1, 2, 3]
my_int = 10

print("Outside function (before calling):")
print("Mutable list:", my_list)
print("Immutable int:", my_int)

modify_arguments(my_list, my_int)

print("Outside function (after calling):")
print("Mutable list:", my_list)
print("Immutable int:", my_int)
```

```python
# Que.8  Write a program to demonstrate the use of logical operators.

x = 10
y = 5

print("Logical AND:")
print(x > 5 and y < 10)  # True and True == True
print(x > 5 and y > 10)  # True and False == False

print("\nLogical OR:")
print(x > 5 or y < 10)   # True or True == True
print(x < 5 or y < 10)   # False or True == True

print("\nLogical NOT:")
print(not(x > 5 and y < 10)) # not(True and True) == not(True) == False
print(not(x < 5 or y < 10))  # not(False or True) == not(True) == False
```

```python
# Que.10 Write code to demonstrate type casting with list elements.

my_list = ["1", 2, 3.14, "True"]

print("Original list:", my_list)
print("Types of elements:", [type(i) for i in my_list])

# type casting list elements
int_element = int(my_list[0])
float_element = float(my_list[1])
string_element = str(my_list[2])
bool_element = bool(my_list[3])


print("\nAfter type casting:")
print("First element as integer:", int_element, type(int_element))
print("Second element as float:", float_element, type(float_element))
print("Third element as string:", string_element, type(string_element))
print("Fourth element as boolean:", bool_element, type(bool_element))


# Example of implicitly casting while performing operations
sum_of_numbers = my_list[1] + my_list[2]
print("\nImplicit type casting (int + float):", sum_of_numbers, type(sum_of_numbers))
```

```python
# Que.13 Write a Python program to find the sum of all even numbers between 1 and 50.

sum_of_even = 0
for i in range(1, 51):
  if i % 2 == 0:
    sum_of_even += i

print("The sum of all even numbers between 1 and 50 is ", sum_of_even)
```

```python
# Que.14 Write a program to reverse a string using a while loop.

string = input("Enter a string: ")
reversed_string = ""
length = len(string)
index = length - 1

while index >= 0:
  reversed_string += string[index]
  index -= 1

print("The reversed string is: ", reversed_string)
```

```python
# Que.15 Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number: "))

if num < 0:
  print("Factorial is not defined for negative numbers.")
elif num == 0:
  print("The factorial of 0 is 1.")
else:
  factorial = 1
  n = num

  while n > 0:
    factorial *= n
    n -= 1
  print(f"The factorial of {num} is {factorial}")
```
