1. What is the difference between a function and a method in Python?
    - Functions are independent, while method is a function that is dependent on an object and defined within a class.
    - Functions are called directly by name, while methods are called on an object instance.
    - Functions typically operate on their input arguments, while methods can access and modify the state of the object they belong to.
    - Functions are ideal for general-purpose tasks and reusable utility code, whereas methods are used to define the behaviors and operations specific to a particular object.

    - **Function Example** -  
    def greet(name): print("Hello", name) greet("Alice")
    
    - **Method Example** -  
    class Greeter: def greet(self, name): print("Hello", name) g = Greeter() g.greet("Alice")

    - Regular functions calculates the values and return in one go. Since its returning in one go, memory usage will be more and it will take a lot of time to execute if there is a lot of computation. Whereas generators produce items one at a time, only when requested. This is crucial for large databases, as it prevents high memory consumption. Generators only hold the current iteration's value in memory, making them very efficient.

8. What is a lambda function in Python and when is it typically used?
    - lambda function in Python is a anonymous function defined using the **lambda** keyword. lambda function do not have a name and it is restricted to a single expression.  
    - Lambda functions are typically used in scenarios where a simple function is required for a short period and does not need to be formally defined with a name. Lambda functions are frequently used with built-in functions like map(), filter(), sorted(), and reduce(), where a simple operation needs to be applied to elements of an iterable.  
    Example - a = [1, 2, 3, 4] list(map(lambda x : x**2, a))

9. Explain the purpose and usage of the `map()` function in Python.
    - The map() function in Python is used to apply a given function to each item of an iterable (such as a list, tuple, or set) and return a map object, which is an iterator containing the results. It provides a concise and efficient way to perform operations on all elements within a collection without explicitly writing a loop.  
    - The map() function takes two main arguments -  
        - **function**: a function that will be applied to each item of the iterable.
        - **iterable**: an iterable whose elements will be passed to the function one by one.  
        Example - x = [3, 4, 5, 6] sq_num = list(map(lambda x : x**2, x)) print(sq_num)

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
    - **map() function -**
        - Applies a given function to each item in an iterable and returns a map object containing the results.  
        **Example -** num = [1, 2, 3, 4] doubled_numbers = list(map(lambda x: x * 2, num)
    - **reduce() function -**
        - Applies a function cumulatively to the items of an iterable, reducing the iterable to a single value. reduce is not a python keyword, you have to import it from a library called functools.  
        **Example -** from functools import reduce num = [1, 2, 3, 4] sum_of_num = reduce(lambda x, y: x + y, num)
    - **filter() function -**
        - It is used to filter elements from an iterable based on some condition.  
        **Example -** numbers = [1, 2, 3, 4, 5] even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

```python
# Que.1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

numbers = [5, 8, 16, 29, 17, 28, 10]
result = sum_even_numbers(numbers)
print(result)
```

```python
# Que.2 Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(string):
    reversed_string = ""
    for i in range(len(string) - 1, -1, -1):
        reversed_string += string[i]
    return reversed_string

string = "Monica"
result = reverse_string(string)
print(result)
```

```python
# Que.3 Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def sq_num(numbers):
    squared_numbers = []
    for num in numbers:
        squared_numbers.append(num ** 2)
    return squared_numbers

sq_num([1, 2, 3, 4, 5])
```

```python
# Que.4 Write a Python function that checks if a given number is prime or not from 1 to 200.

import math
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num)
```

```python
# Que.5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci_sequence(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b

for num in fibonacci_sequence(10):
    print(num)
```

```python
# Que.6 Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for num in power_of_two(5):
    print(num)
```

```python
# Que.7 Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

with open("my_file.txt", "w") as f:
    f.write("This is line 1\n")
    f.write("This is line 2\n")
    f.write("This is line 3\n")

for line in read_file_lines("my_file.txt"):
    print(line.strip())
```

```python
# Que.8 Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

lst = [('apple', 90), ('banana', 50), ('cherry', 120), ('date', 250), ('mango', 80)]
sorted_lst = sorted(lst, key=lambda x: x[1])
print(sorted_lst)
```

```python
# Que.9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

temp_in_cels = [8, 4, 16, 28, 19]
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
temp_in_fahr = list(map(celsius_to_fahrenheit, temp_in_cels))
print(temp_in_fahr)
```

```python
# Que.10 Create a Python program that uses `filter()` to remove all the vowels from a given string.

string = "Nandkishore"
def remove_vowels(string):
    vowels = ['a', 'e', 'i', 'o', 'u']
    return ''.join([char for char in string if char.lower() not in vowels])

print(remove_vowels(string))
```
